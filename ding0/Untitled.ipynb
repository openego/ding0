```python
import geopandas as gpd
from shapely import wkb
from requests import get

http_proxy = "http://intranet.ise.fhg.de:3128"
https_proxy = "https://intranet.ise.fhg.de:3128"

proxies = {
    "http": http_proxy,
    "https": https_proxy,
}

version = 'v0.4.5' # latest run eGo Dataprocessing
# today: 22.05.04

# all hs/ms substation
hvmv_subst = get('https://openenergy-platform.org/api/v0/schema/grid/tables/ego_dp_hvmv_substation/rows/?where=version='+version, verify=False)  # , proxies=proxies
gdf_hvmv_subst = gpd.GeoDataFrame(hvmv_subst.json())
gdf_hvmv_subst['geom'] = gdf_hvmv_subst['geom'].apply(lambda x: wkb.loads(x, hex=True))
gdf_hvmv_subst = gdf_hvmv_subst.set_geometry('geom')
gdf_hvmv_subst.crs = 'epsg:3035'
```

```python
%%time

gdf_hvmv_subst['cordinates_lon_lat'] = gdf_hvmv_subst.apply(lambda x: [[x.lon, x.lat]], axis=1)
gdf_hvmv_subst['cordinates_lat_lon'] = gdf_hvmv_subst.apply(lambda x: [[x.lat, x.lon]], axis=1)
```
